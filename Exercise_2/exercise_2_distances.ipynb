```python
import pandas as pd
from scipy.spatial import distance

# Load dataset
df = pd.read_csv("input_points.csv")

# Calculate distances
results = []
for i, row in df.iterrows():
    p1 = [row['Point1_x'], row['Point1_y']]
    p2 = [row['Point2_x'], row['Point2_y']]
    euclidean = distance.euclidean(p1, p2)
    manhattan = distance.cityblock(p1, p2)
    minkowski_3 = distance.minkowski(p1, p2, p=3)
    results.append({
        'Index': i,
        'Point1': p1,
        'Point2': p2,
        'Euclidean': euclidean,
        'Manhattan': manhattan,
        'Minkowski (p=3)': minkowski_3
    })

distances_df = pd.DataFrame(results)
print(distances_df.head())
```
